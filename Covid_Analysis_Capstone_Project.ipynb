1.	Import the dataset using Pandas from above mentioned url.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

Url = 'https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv'

covid_19 = pd.read_csv(Url)

covid_19.head()
```

3.	Low Level Data Understanding :
    a.	Find count of unique values in location column.
    b.	Find which continent has maximum frequency using values counts.
    c.	Find maximum & mean value in 'total_cases'.
    d.	Find 25%,50% & 75% quartile value in 'total_deaths'.
    e.	Find which continent has maximum 'human_development_index'.
    f.	Find which continent has minimum 'gdp_per_capita'.


3. a.	Find count of unique values in location column.

```python
item_counts = covid_19["location"].value_counts()
print('Count of unique values:')
print(item_counts)
```

3. b.	Find which continent has maximum frequency using values counts.

```python
max_frq  = covid_19.continent.value_counts().max()
print('continent has maximum frequency: \n',max_frq)
```

  3. c.	Find maximum & mean value in 'total_cases'.

```python
max_element = covid_19['total_cases'].max().round(2)
print('Maximum in total cases column:\n',max_element)

mean_element = covid_19['total_cases'].mean().round(2)
print('Mean in total cases column:\n',mean_element )
```

3. d.	Find 25%,50% & 75% quartile value in 'total_deaths'.

```python
percentile_25 = covid_19.total_deaths.quantile(0.25)
print(percentile_25)
percentile_50 = covid_19.total_deaths.quantile(0.50)
print(percentile_50)
percentile_75 = covid_19.total_deaths.quantile(0.75)
print(percentile_75)
```

3. e.	Find which continent has maximum 'human_development_index'.

```python
max_HDI = covid_19['human_development_index'].max().round(3)
print('Continent has maximum in human_development_index: \n',max_HDI)
```

3. f.	Find which continent has minimum 'gdp_per_capita'.

```python
min_GDP = covid_19['gdp_per_capita'].min().round(3)
print('Continent has minimum gdp_per_capita : ')
print(min_GDP)
```

4.	Filter the dataframe with only this columns ['continent','location','date','total_cases','total_deaths','gdp_per_ca
pita','human_development_index'] and update the data frame.


```python
new_covid19 = covid_19.filter(['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'], axis="columns")
covid_19.update(new_covid19)
print(new_covid19)
```

6.	Date time format :
    a.	Convert date column in datetime format using pandas.to_datetime
 
    b.	Create new column month after extracting month data from date column.


6. a.	Convert date column in datetime format using pandas.to_datetime

```python
new_covid19['date'] = pd.to_datetime(new_covid19['date'])
new_covid19.head()
```

6. b.	Create new column month after extracting month data from date column.

```python
new_covid19['month'] = new_covid19['date'].dt.month
new_covid19.head()
```

8.	Feature Engineering :
    a.	Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'


```python
new_covid19["total_deaths_to_total_cases"] = (new_covid19["total_deaths"] / new_covid19["total_cases"])
new_covid19.head()
```
